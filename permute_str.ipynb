```python
def permuteHelper(s, chosen):
    print 'permute("{}", "{}")'.format(s, chosen)
    if not s:
        print chosen
    else:
        for i in xrange(len(s)):
            c = s[i]
            chosen += c
            left = s[:i] + s[i+1:]
            permuteHelper(left, chosen)
            chosen = chosen[-2:]

def permute(s):
    permuteHelper(s, '')
            
```
