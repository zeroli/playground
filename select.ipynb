```python
def _min(heap, cj1, cj2):
    if cj1 >= len(heap): return cj2
    if cj2 >= len(heap): return cj1
    if heap[cj1] < heap[cj2]:
        return cj1
    else:
        return cj2

def _max(heap, cj1, cj2):
    if cj1 >= len(heap): return cj2
    if cj2 >= len(heap): return cj1
    if heap[cj1] > heap[cj2]:
        return cj1
    else:
        return cj2

def _max_downbubble(heap, j):
    while True: # down bubbling for each subtree
        cj1, cj2 = 2*j+1, 2*j+2
        #import pdb; pdb.set_trace()
        jj = _max(heap, cj1, cj2)
        if jj < len(heap) and heap[jj] > heap[j]:
            heap[jj], heap[j] = heap[j], heap[jj]
            j = jj
        else: break

def _min_downbubble(heap, j):
    while True: # down bubbling for each subtree
        cj1, cj2 = 2*j+1, 2*j+2
        #import pdb; pdb.set_trace()
        jj = _min(heap, cj1, cj2)
        if jj < len(heap) and heap[jj] < heap[j]:
            heap[jj], heap[j] = heap[j], heap[jj]
            j = jj
        else: break

def _build_max_heap(l):
    heap = l[:]
    print 'build max heap from {}'.format(heap)
    size = len(heap)
    i = (size-1-1)/2
    while i >= 0:
        _max_downbubble(heap, i)
        print i, heap
        i -= 1
    print 'max-heap: ', heap
    return heap

def _build_min_heap(l):
    heap = l[:]
    print 'build min heap from {}'.format(heap)
    size = len(heap)
    i = (size-1-1)/2
    while i >= 0:
        _min_downbubble(heap, i)
        print i, heap
        i -= 1
    print 'min-heap: ', heap
    return heap

def _max_heap_remove(heap):
    print 'remove max from heap {}'.format(heap)
    x = heap[0]
    heap[0] = heap.pop()
    _max_downbubble(heap, 0)
    return x

def _min_heap_remove(heap):
    print 'remove min from heap {}'.format(heap)
    x = heap[0]
    heap[0] = heap.pop()
    _min_downbubble(heap, 0)
    return x
```

```python
def heapselect1(L, K):
    """
    select K smallest items from list L
    algorithm will build one max-heap using K items, and then insert the remaining items
    into this heap, but finally only keep K items in this K max-heap
    """
    print 'algo1: select {} smallest items from {}'.format(K, L)
    if len(L) <= K: return L
    max_heap = _build_max_heap(L[0:K])
    for x in L[K:]:
        if x > max_heap[0]:
            continue
        else:
            max_heap[0] = x
            _max_downbubble(max_heap, 0)
    print ">> ", max_heap
    return max_heap

def heapselect2(L, K):
    """
    select K smallest items from list L
    algorithm will build one min-heap using the whole list, and then select min item from this heap K items,
    to get final K smallest items
    """
    print 'algo2: select {} smallest items from {}'.format(K, L)
    if len(L) <= K: return L
    min_heap = _build_min_heap(L)
    Kitems = []
    for i in range(K):
        Kitems.append(_min_heap_remove(min_heap))
    print ">> ", Kitems
    return Kitems

def _partition(L, left, right, pivot):
    ii, jj = left, left
    pv = L[pivot]
    L[pivot], L[right] = L[right], L[pivot]    
    while jj <= right:
        if L[jj] < pv:
            L[jj], L[ii] = L[ii], L[jj]
            ii += 1
        jj += 1
    L[right], L[ii] = L[ii], L[right]
    return ii
    
def quickselect_recursive(L, K):
    """
    select K smallest items from list L
    algorithm will be based on quickselection with recursive method
    """
    print 'algo3: select {} smallest items from {}'.format(K, L)
    if len(L) <= K: return L
    
    def _select(L, left, right, K):
        pivot = (left+right)/2
        print '1: pivot {}: {}'.format(pivot, L)
        pivot = _partition(L, left, right, pivot)
        print '2: pivot {}: {}'.format(pivot, L)
        if pivot == K:
            return L[:K]
        elif K < pivot:
            return _select(L, left, pivot-1, K)
        else:
            return _select(L, pivot+1, right, K)
        
    Kitems = _select(L[:], 0, len(L)-1, K)
    print " >> ", Kitems
    return Kitems

def quickselect_iterative(L, K):
    """
    select K smallest items from list L
    algorithm will be based on quickselection with iterativev method
    """
    print 'algo4: select {} smallest items from {}'.format(K, L)
    if len(L) <= K: return L
    
    def _select(L, left, right, K):
        while left <= right:
            pivot = (left+right)/2
            print '1: pivot {}: {}'.format(pivot, L)
            pivot = _partition(L, left, right, pivot)
            print '2: pivot {}: {}'.format(pivot, L)
            if pivot == K:
                return L[:K]
            elif K < pivot:
                right = pivot - 1
            else:
                left = pivot + 1
        
    Kitems = _select(L[:], 0, len(L)-1, K)
    print " >> ", Kitems
    return Kitems
```

```python
L = [random.randint(0, 100) for x in xrange(11)]
K = 8
select1 = heapselect1(L, K)
assert set(select1) == set(sorted(L)[0:K])

select2 = heapselect2(L, K)
assert set(select2) == set(sorted(L)[0:K])

select3 = quickselect_recursive(L, K)
assert set(select3) == set(sorted(L)[0:K])

select4 = quickselect_iterative(L, K)
assert set(select4) == set(sorted(L)[0:K])
```
